```python
# Assuming 'authors' and 'titles' are your pandas DataFrames
combined_df = pd.concat([authors, titles], axis=1)

# Drop duplicates where both 'authors' and 'titles' are the same
unique_df = combined_df.drop_duplicates()

print("Data after dropping duplicates in both authors and titles:")
print(unique_df)

# Split back into 'authors' and 'titles'
authors_unique = unique_df['Book-Author']
titles_unique = unique_df['Book-Title']

print("\nAuthors after dropping duplicates:")
print(authors_unique)

print("\nTitles after dropping duplicates:")
print(titles_unique)
```

```python
authors, titles = authors_unique, titles_unique
```

```python
from sentence_transformers import SentenceTransformer
```

```python
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
```

```python
titles[0]
```

```python
title_list = [title for title in titles]
author_list = [author for author in authors]
```

```python
author_list[0]
```

```python
concat_list = []

for idx in range(len(title_list)):
  concat_list.append(str(title_list[idx]) + ' ' + str(author_list[idx]))
```

```python
concat_list
```

```python
# titles_embedded = model.encode(title_list)
# author_embedded = model.encode(author_list)
concat_embedded = model.encode(concat_list)
```

```python
titles.head()
```

```python
len(titles.values)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def similar_embeddings(input, k=10):
    '''
        input: string
        return: list of similar embeddings
    '''
    print('[+] Retrieving similar embeddings...')
    input_embedding = model.encode(input)
    # use cosine similarity to find similar embedding
    similarity_list = []
    # calculate similarity score for each pair of <input, embedding, index>
    for i, embedding in enumerate(concat_embedded):
        similarity = cosine_similarity([input_embedding], [embedding])[0][0]
        similarity_list.append((similarity,  titles.values[i]))
        # print(f'Title {i}',titles.values[i])
    # get top 10
    similarity_list = sorted(similarity_list, reverse=True)[:k]
    # return embeddings with highest similarity score
    return similarity_list
```

```python
print(similar_embeddings("Clara Callan"))
```
